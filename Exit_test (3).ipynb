Encoding

```python
#encode this column using label encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Attrition']=le.fit_transform(df['Attrition'])
```

```python
frequency_enc=["Gender","Job Role","Work-Life Balance","Job Satisfaction",
               "Performance Rating","Overtime","Education Level","Marital Status",
               "Job Level","Company Size","Remote Work","Leadership Opportunities",
               "Innovation Opportunities","Company Reputation","Employee Recognition"]
```

```python
# Apply Frequency Encoding
for col in frequency_enc:
    # Compute frequency of each category in the column
    frequency_map = df[col].value_counts(normalize=True)

    # Map these frequencies to the original column
    df[col] = df[col].map(frequency_map)
```

#Selecting features and target columns

```python

#Selecting features and target columns


x=df.drop(columns=['Attrition'])
y=df['Attrition']
```

#spliting the dataset into train and test

```python
#spliting the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

Model applying & Hyperparameters

```python
#select feature using RFE and apply random forest method


from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest Classifier
rf=RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    bootstrap=True,
    random_state=42
)

#create a logistic regression with hyperparameters
lg=LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0,
                       fit_intercept=True, intercept_scaling=1, class_weight=None,
                       random_state=None, solver='lbfgs', max_iter=100,
                       verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)




rfe=RFE(estimator=lg,n_features_to_select=10)
rfe=RFE(estimator=rf,n_features_to_select=10)
x_train_rfe=rfe.fit_transform(x_train,y_train)
x_test_rfe=rfe.transform(x_test)
```

```python
# Print selected features
selected_features = rfe.support_
feature_names = df.drop(columns=['Attrition']).columns
selected_feature_names = feature_names[selected_features]
print("Selected Features for Prediction:", selected_feature_names)

```

#scaling

```python
#scaling
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train_rfe)
x_test_scaled=sc.transform(x_test_rfe)
```

```python
#train the model

lg.fit(x_train_scaled,y_train)
```

```python
rf.fit(x_train_scaled,y_train)
```

#predict the model

```python
#predict the model
y_pred=lg.predict(x_test_scaled)
```

```python
y_pred
```

```python
#predict the model
y_pred=rf.predict(x_test_scaled)
y_pred
```

Evaluation

```python
from sklearn.metrics import accuracy_score


accuracy=lg.score(x_test_scaled,y_test)
print(f'Accuracy with reduced features: {accuracy:.2f}')
```

```python
from sklearn.metrics import accuracy_score


accuracy=rf.score(x_test_scaled,y_test)
print(f'Accuracy with reduced features: {accuracy:.2f}')
```
